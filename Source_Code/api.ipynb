```python
import requests

def fetch_sentiment(text):
    # Replace with your actual API key
    api_key = 'AIzaSyBpLetJX8VwYq3QHc_icj9po2DWiQRqMdY'
    url = f"https://generativelanguage.googleapis.com/v1/models/gemini-1.5-flash:generateContent?key={api_key}"

    # Sentiment analysis prompt tailored for BBC articles
    payload = {
        "contents": [
            {
                "role": "user",
                "parts": [
                    {
                        "text": (
                            "You are a sentiment analysis model trained to assess the emotional tone of BBC news articles across diverse categories. "
                            "Carefully analyze the text and classify it into one of the following categories:\n"
                            "1. 'Positive'\n"
                            "2. 'Negative'\n"
                            "3. 'Neutral'\n"
                            "Return only one of these exact words: 'Positive', 'Negative', or 'Neutral' without any additional explanation."
                        )
                    }
                ]
            },
            {
                "role": "user",
                "parts": [{"text": text}]
            }
        ]
    }

    headers = {"Content-Type": "application/json"}

    try:
        response = requests.post(url, json=payload, headers=headers)
        response.raise_for_status()
        result = response.json()

        classification = result["candidates"][0]["content"]["parts"][0]["text"].strip()

        if classification in ["Positive", "Negative", "Neutral"]:
            return classification
        else:
            return "Unknown Sentiment"

    except requests.exceptions.RequestException as e:
        print("API request error:", e)
        return "API Error"

    except (KeyError, IndexError):
        print("Error parsing response.")
        return "Parse Error"
```

```python
while True:
    text=input("Enter the Text:")
    sentiment = fetch_sentiment(text)
    print('\nPredicted =======>',str(sentiment)+"\n\n")
```
